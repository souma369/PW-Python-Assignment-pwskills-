```python
#The pyplot module of Matplotlib is a powerful tool for creating various types of plots. Here are five common types of plots that can be created using pyplot:

 #   1. Line Plot:

#A line plot is used to visualize the relationship between two continuous variables. It's often used to display trends or patterns over time.


import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Line Plot Example')
plt.show()

 #  2. Scatter Plot:

# A scatter plot is used to display the relationship between two variables, typically showing individual data points.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Scatter Plot Example')
plt.show()

#   3. Bar Plot:

#A bar plot (or bar chart) is used to represent categorical data with rectangular bars. It's useful for comparing different categories.

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 9]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()

#  4. Histogram:

# A histogram is used to visualize the distribution of a dataset. It shows the frequency of different values within a range.

import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # Generate random data

plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()

# 5. Pie Chart:

#A pie chart is used to represent the proportions of different categories in a dataset. It's especially effective when showing parts of a whole.

import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()





```

## What is a scatter plot?

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x= 3 + np.random.normal(0,2,50)
y= 3 + np.random.normal(0,2,len(x))
```

```python
plt.xlabel("along x axis values")
plt.ylabel("along y axis values")
plt.title("This is X vs Y graph")
colours=np.random.rand(50)
sizes=500*np.random.rand(50)
plt.scatter(x,y, c=colours, s=sizes,alpha=1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.title("company with profit")
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.bar(company,profit)
plt.show()
```

```python
plt.title("company with profit")
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.barh(company,profit)
plt.show()
```
